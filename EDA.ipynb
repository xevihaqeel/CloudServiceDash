Next I'll be correcting the data    

```python
#filing all nan values with unknown
car_ad = car_ad.fillna('Unknown')

#change the model year values to ints
car_ad['model_year'] = car_ad['model_year'].apply(lambda x: int(x) if x != 'Unknown' else x)

#changing odemeter values to ints
car_ad['odometer'] = car_ad['odometer'].apply(lambda x: int(x) if x != 'Unknown' else x)

#correcting the values in the is_4wd to read yes and no instead of 1.0 and Unknown
car_ad['is_4wd'] = car_ad['is_4wd'].replace({1.0: 'yes', 'Unknown': 'no'})

car_ad['cylinders'] = car_ad['cylinders'].apply(lambda x: int(x) if x != 'Unknown' else x)
```
